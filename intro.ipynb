# ðŸŸ¦ **1. Dimensional Overview (1D â†’ 5D)**

| Dimension | Shape Example | Meaning | Mental Model |
|----------|----------------|---------|---------------|
| **1D** | `(n,)` | A list of numbers | Line |
| **2D** | `(rows, cols)` | Matrix / table | Spreadsheet |
| **3D** | `(blocks, rows, cols)` | Stack of matrices | Book of pages |
| **4D** | `(groups, blocks, rows, cols)` | Stack of 3D arrays | Bookshelf of books |
| **5D** | `(sets, groups, blocks, rows, cols)` | Stack of 4D arrays | Library of shelves |

# ðŸŸ© **2. Indexing Cheat Sheet (1D â†’ 5D)**

## **1D Indexing**
```python
arr[i]
```

## **2D Indexing**
```python
arr[row, col]
arr[row][col]
```

## **3D Indexing**
Shape: `(B, R, C)`
```python
arr[block, row, col]
```

Examples:
```python
arr[0]        # block 0 â†’ shape (R, C)
arr[0, 1]     # row 1 of block 0 â†’ shape (C,)
arr[0, 1, 2]  # element
```

## **4D Indexing**
Shape: `(G, B, R, C)`
```python
arr[group, block, row, col]
```

Examples:
```python
arr[1]            # group 1 â†’ shape (B, R, C)
arr[1, 0]         # block 0 of group 1 â†’ shape (R, C)
arr[1, 0, 1]      # row 1 â†’ shape (C,)
arr[1, 0, 1, 3]   # element
```

## **5D Indexing**
Shape: `(S, G, B, R, C)`
```python
arr[set, group, block, row, col]
```

Examples:
```python
arr[0]               # set 0 â†’ shape (G, B, R, C)
arr[0, 1]            # group 1 â†’ shape (B, R, C)
arr[0, 1, 2]         # block 2 â†’ shape (R, C)
arr[0, 1, 2, 0]      # row 0 â†’ shape (C,)
arr[0, 1, 2, 0, 3]   # element
```

# ðŸŸ¨ **4. NumPy â†’ PyTorch Tensor Mapping**

| Concept | NumPy | PyTorch |
|--------|--------|----------|
| Create array/tensor | `np.array([...])` | `torch.tensor([...])` |
| Shape | `arr.shape` | `tensor.shape` |
| Reshape | `arr.reshape(...)` | `tensor.reshape(...)` |
| Add dimension | `np.expand_dims(arr, axis)` | `tensor.unsqueeze(axis)` |
| Remove dimension | `np.squeeze(arr)` | `tensor.squeeze()` |
| Transpose | `arr.transpose(...)` | `tensor.permute(...)` |
| 4D image batch | `(N, H, W, C)` | `(N, C, H, W)` |
| 5D video batch | `(N, T, H, W, C)` | `(N, C, T, H, W)` |

**Key difference:**  
PyTorch uses **channelsâ€‘first**, NumPy uses **channelsâ€‘last** by default.

# ðŸŸª **5. Reshape Rules (Critical)**

### âœ” Total number of elements must stay the same
```python
arr.size == np.prod(new_shape)
```

### âœ” NumPy fills rowâ€‘major (Câ€‘order)
Meaning: last dimension changes fastest.

### âœ” Use `-1` to autoâ€‘infer
```python

```

# ðŸŸ« **6. Slicing Patterns (All Dimensions)**

### **General rule**
```
arr[a:b, c:d, e:f, ...]
```

### Examples

#### 3D
```python
arr[:, 0]      # all blocks, row 0
arr[1, :, :]   # block 1, all rows/cols
```

#### 4D
```python
arr[:, 0, :, :]   # all groups, block 0
arr[1, :, :, 2]   # group 1, all blocks/rows, column 2
```

#### 5D
```python
arr[:, :, 0]      # all sets, all groups, block 0
```
